```python
import numpy as np
import matplotlib.pyplot as plt

# continuous function actual value
def f(x):
    return 1.60934 * x

# discrete fibonacci sequence for estimation
fib = []
for i in range(100):
    if i < 2:
        fib.append(1)
    else:
        fib.append(fib[-1] + fib[-2])
```

```python
actual = np.array(fib)
estimation = f(np.array([1] + fib))[:-1]
diff = np.abs(estimation - actual) / actual * 100
```

```python
# fitst 10 values
limit = 10
y_act = actual[:limit]
y_est = estimation[:limit]
x = range(limit)
plt.title('Fibonacci versus Formula')
plt.plot(x, y_act, 'g')
plt.plot(x, y_est, 'b')
plt.axis([x[0], x[-1], min(y_act), max(y_act)])
plt.show()
```

```python
limit = 90
y_act = actual[:limit]
y_est = estimation[:limit]
x = range(limit)
plt.title('Fibonacci versus Formula')
plt.plot(x, y_act, 'g')
plt.plot(x, y_est, 'b')
plt.axis([x[0], x[-1], min(y_act), max(y_act)])
plt.show()
```

```python
limit = 90
y = diff[:limit]
x = range(limit)
plt.title('Error Between Fibonacci and Formula')
plt.plot(x, y, 'r')
plt.axis([x[0], x[-1], min(y), max(y)])
plt.show()
```
